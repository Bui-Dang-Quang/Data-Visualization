```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#--------------------------#
#1. Contour plot and 2. Heatmap
#--------------------------#
# Read text file
with open('Homework_sampledata1.txt','r') as f:
    lines = f.readlines() #Read all lines


print ('---------------------------------------------------------------------------------------------')
line_array = []
for i in range(1, len(lines)):
    value = lines[i].split()
    row = []
    for v in value[1:]:
        v = float(v)
        if v != -99.0:
            row.append(v)
        else:
            row.append(None)
    line_array.append(row)
print(line_array)
print ('---------------------------------------------------------------------------------------------')
line_array2 = np.transpose(line_array)
data_raw = []
year = 2008
for month, row in enumerate (line_array2, start=1):
     for day, value in enumerate(row, start=1):
        print(f"{year}-{month:02d}-{day:02d} 00:00:00 {value}")
        data_raw.append([f"{year}-{month:02d}-{day:02d}", "00:00:00", value])

df = pd.DataFrame(data_raw, columns=['Date', 'Time', 'Temperature'])
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Assuming df is already defined and includes 'Date' and 'Temperature' columns
df['Month'] = pd.to_datetime(df['Date']).dt.month
df['Day'] = pd.to_datetime(df['Date']).dt.day

# Pivot the data to create a 2D grid (Days as rows, Months as columns)
pivot_df = df.pivot(index='Month', columns='Day', values='Temperature')

# Create the x and y mesh grid for contour plotting
X, Y = np.meshgrid(pivot_df.columns, pivot_df.index)

# Define the color boundaries and levels for the color mapping
bounds = np.linspace(0, 32, 17)  # 0 to 32 with 16 intervals (0, 2, ..., 32)
cmap = plt.cm.jet

# Plot the contour plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 15))

# Contour plot with specified levels
contour = ax1.contourf(X, Y, pivot_df.values, cmap=cmap, levels=bounds)
cbar1 = fig.colorbar(contour, ax=ax1, extend='both')  # Add colorbar and extend for better visualization
cbar1.set_ticks(bounds)  # Set colorbar ticks based on the defined bounds
cbar1.ax.tick_params(labelsize=17)  # Set font size for colorbar ticks

# Set labels and title
ax1.set_xlabel('Day', fontsize=17)
ax1.set_yticks(np.arange(1, 13))
ax1.set_ylabel('Month', fontsize=17)
ax1.set_title('Temperature Contour Plot', fontsize=17)

# Set font size for tick labels
ax1.tick_params(axis='both', which='major', labelsize=17)

# Heat map plot with specified color scale
heatmap = ax2.pcolormesh(X, Y, pivot_df.values, cmap=cmap, norm=mpl.colors.BoundaryNorm(bounds, cmap.N), shading='auto')
# Add colorbar to the heatmap and extend
cbar2 = fig.colorbar(heatmap, ax=ax2, extend='both')
cbar2.set_ticks(bounds)  # Set colorbar ticks based on the defined bounds
cbar2.ax.tick_params(labelsize=17)  # Set font size for colorbar ticks

# Set labels and title
ax2.set_xlabel('Day', fontsize=17)
ax2.set_ylabel('Month', fontsize=17)
ax2.set_title('Temperature Heatmap', fontsize=17)

# Set the x and y ticks to reflect the mesh grid
ax2.set_xticks(np.arange(X.min(), X.max() + 1, 1))  # Set appropriate intervals for day
ax2.set_yticks(np.arange(Y.min(), Y.max() + 1, 1))  # Set appropriate intervals for month

# Set font size for tick labels
ax2.tick_params(axis='both', which='major', labelsize=17)

# Increase space between 2 subplots
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
from math import pi
import matplotlib.pyplot as plt
import numpy as np

# Assuming you've already calculated the means (Feature_1 to Feature_6)
features = [Feature_1, Feature_2, Feature_3, Feature_4, Feature_5, Feature_6]
labels = ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5', 'Feature 6']

# Number of features
num_vars = len(features)

# Compute angle for each axis (in radians)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot should be a closed loop, so append the first value at the end
features += features[:1]
angles += angles[:1]

# Initialize the radar plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw the radar chart
ax.fill(angles, features, color='blue', alpha=0.25)
ax.plot(angles, features, color='blue', linewidth=2)

# Set labels for each axis
ax.set_yticklabels([])  # Hide default radial labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Custom radial ticks (set the circle benchmarks to 10, 20, 30, etc.)
ax.set_ylim(0, 40)  # Set the limit to 30, the highest scale
ax.set_yticks([10, 20, 30])  # Add custom tick marks
ax.set_yticklabels(['10', '20', '30'], color="grey", size=12)  # Label them

# Set the radial labels to align along the line between Feature 4 and Feature 1
# Find the angle of the axis for Feature 4 (since we want the radial ticks along that axis)
angle_feature4 = angles[0] * 180 / np.pi  # Convert from radians to degrees
ax.set_rlabel_position(angle_feature4)  # Set the radial tick label position

# Set offset and direction for the radar plot (as in your example)
ax.set_theta_offset(4 * pi / 2)
ax.set_theta_direction(1)

plt.show()
```

```python
# Initialize dictionary to store features
monthly_features = {}

# Function to calculate the mean, excluding NaN values
def calculate_mean_with_nan(month_data, start, end):
    valid_values = month_data.loc[start:end, 'Temperature'].dropna()
    return valid_values.mean() if not valid_values.empty else np.nan

# Loop through each month to calculate the features
for month in range(1, 13):
    month_data = df[df['Month'] == month].reset_index(drop=True)
    
    # Calculate the means for each feature
    C1 = calculate_mean_with_nan(month_data, 0, 4)   # Days 1-5
    C2 = calculate_mean_with_nan(month_data, 5, 9)   # Days 6-10
    C3 = calculate_mean_with_nan(month_data, 10, 14) # Days 11-15
    C4 = calculate_mean_with_nan(month_data, 15, 19) # Days 16-20
    C5 = calculate_mean_with_nan(month_data, 20, 24) # Days 21-25
    
    # Handle the different number of days per month
    if month in [1, 3, 5, 7, 8, 10, 12]:  # For 31-day months
        C6 = calculate_mean_with_nan(month_data, 25, 30)  # Days 26-31
    elif month in [4, 6, 9, 11]:  # For 30-day months
        C6 = calculate_mean_with_nan(month_data, 25, 29)  # Days 26-30
    else:  # February (28 or 29 days)
        C6 = calculate_mean_with_nan(month_data, 25, 28)  # Days 26-28
    
    # Store the features for the month
    monthly_features[month] = [C1, C2, C3, C4, C5, C6]

# Convert
mean_5days = np.array(list(monthly_features.values()))


# Define a 3x4 grid for the subplots
fig, axes = plt.subplots(3, 4, figsize=(15, 15), subplot_kw=dict(polar=True))
fig.suptitle('Radar Plots of average temperature of each 5 days per month in 2008', fontsize=16)

# Define labels for the features
labels = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6']
num_vars = len(labels)

# Loop through each month and plot the radar chart
for i, (month, features) in enumerate(monthly_features.items()):
    # Select the appropriate subplot
    ax = axes[i // 4, i % 4]
    
    # Prepare the data for radar plot
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    features += features[:1]  # Close the plot
    angles += angles[:1]

    # Draw the radar chart
    ax.fill(angles, features, color='blue', alpha=0.25)
    ax.plot(angles, features, color='blue', linewidth=2)

    # Set labels for each axis
    ax.set_yticklabels([])  # Hide default radial labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    # Set custom radial ticks (circle benchmarks)
    ax.set_ylim(0, 40)  # Set the limit to 40
    ax.set_yticks([10, 20, 30])  # Add custom tick marks
    ax.set_yticklabels(['10', '20', '30'], color="grey", size=8)

    # Find the angle of the axis for Feature 4 (for radial label alignment)
    angle_feature4 = angles[0] * 180 / np.pi  # Convert from radians to degrees
    ax.set_rlabel_position(angle_feature4)  # Set the radial tick label position

    # Set title of the subplot as the month number
    ax.set_title(f'Month {month}', size=14, color='black')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the main title
plt.show()
```

```python
#--------------------------#
#4. Dendrogram
#--------------------------#
from scipy.cluster.hierarchy import dendrogram, linkage

# Create labels for each month
list_month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Perform hierarchical/agglomerative clustering
Z = linkage(mean_5days, 'ward')

# Plot the dendrogram
plt.figure(figsize=(12, 5))
dendrogram(Z, labels=list_month)
plt.title('Dendrogram of Average Temperature Every 5 Days in Each Month of 2008', size=16)
plt.xlabel('Month')
plt.show()
```
