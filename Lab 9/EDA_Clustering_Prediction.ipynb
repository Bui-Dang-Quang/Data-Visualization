# <b> Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# <b> Load data

```python
df = pd.read_csv("SpotifyFeatures.csv")
df.head(5)
```

| **Column Name**     | **Description**                                                                                  |
|----------------------|--------------------------------------------------------------------------------------------------|
| **Genre**           | The musical genre or category the track belongs to. Example: "Movie" for film soundtracks.       |
| **Artist Name**     | The name of the artist or band that performed the track. Example: "Henri Salvador."              |
| **Track Name**      | The title of the song or musical piece. Example: "C'est beau de faire un Show."                  |
| **Track ID**        | A unique identifier for the track, often used for cataloging in streaming services.              |
| **Popularity**      | A numerical value (0–100) representing the track's popularity based on factors like play count.   |
| **Acousticness**    | A score (0.0 to 1.0) indicating how acoustic the track is. Higher values suggest acoustic elements.|
| **Danceability**    | A score (0.0 to 1.0) describing how suitable the track is for dancing.                           |
| **Duration (ms)**   | The length of the track in milliseconds. Example: "99373" equals ~99 seconds.                    |
| **Energy**          | A score (0.0 to 1.0) measuring the intensity and activity of the track.                          |
| **Instrumentalness**| A score (0.0 to 1.0) indicating the likelihood of no vocals. Higher values suggest instrumentals. |
| **Key**             | The musical key of the track, represented as a note (e.g., "C#", "F#").                          |
| **Liveness**        | A score (0.0 to 1.0) reflecting the likelihood that the track was recorded live.                  |
| **Loudness**        | The overall loudness of the track in decibels (dB). Higher values indicate louder tracks.         |
| **Mode**            | Indicates whether the track is in a major (1) or minor (0) key.                                  |
| **Speechiness**     | A score (0.0 to 1.0) measuring the presence of spoken words in the track.                         |
| **Tempo**           | The estimated speed of the track in beats per minute (BPM). Example: "166.969" is fast.          |
| **Time Signature**  | The number of beats per measure. Example: "4/4" is common in Western music.                      |
| **Valence**         | A score (0.0 to 1.0) indicating the musical positiveness or happiness conveyed by the track.      |

# <b> EDA

```python
most_popular = df[df['popularity'] >= 90].sort_values('popularity', ascending=False)
most_popular.head(20)
```

```python
# Check duplicates in the "track_name" column
duplicates = df.duplicated().sum()
print("Number of duplicate:", duplicates)
```

```python
df["duration"]= df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace=True, axis=1)
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
numerical_columns = df.select_dtypes(include=['number']).columns

print("Categorical Columns:", categorical_columns)
print("Numerical Columns:", numerical_columns)
print("\n")
print("Categorical Columns:", len(categorical_columns))
print("Numerical Columns:", len(numerical_columns))
```

```python
numerical_df = df.select_dtypes(include=['number'])
numerical_df
```

```python
corr = numerical_df.corr()
# Set the figure size (width, height)
plt.figure(figsize=(20, 15))  # Adjust the values as needed
sns.heatmap(corr, cmap='YlGnBu', annot=True)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = numerical_df, y= "loudness", x = "energy", color = "c").set(title="loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = numerical_df, y= "popularity", x = "acousticness", color = "b").set(title="popularity vs acousticness")
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot (y='genre', x='duration', data=df, palette="rocket")
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous=df.sort_values("popularity", ascending = False).head(10)
sns.barplot(y= 'genre', x='popularity', data = famous).set(title= "Top 5 Genres by Popularity")
```
